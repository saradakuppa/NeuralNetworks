```python
# Import the Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
# Get the data sets
train = pd.read_csv('HousePricesTrain.csv')
```

```python
test = pd.read_csv('HousePricesTest.csv')
```

```python
# Combine training and test data to perform data cleaning and manupulation 
# But first create a dependent variable which takes the SalePrice from the training set. 
dep_var = train.SalePrice
```

```python
# Drop the SalePrice from training set so that test and training data sets have equal number of columns. 
train.drop(['SalePrice'], axis = 1, inplace = True)
```

```python
combined_data = train.append(test)
```

```python
combined_data.reset_index(inplace=True)
```

```python
# Lets find the columns numerical and non-numerical columns with no nulls
def get_col_with_no_na(df, col_type):
    '''
    col_type:
        num: to get numerical columns with no nulls
        no_num: to get non-numerical columns with no nulls
        all: to get columns with no nulls
    '''
    if(col_type == 'num'):
        predictors = df.select_dtypes(exclude=['object'])
    elif(col_type == 'no_num'):
        predictors = df.select_dtypes(include=['object'])
    elif(col_type == 'all'):
        predictors = df
    else:
        print('Error: choose a type(num, no_num, all)')
        return 0
    cols_with_no_nulls = []
    for col in predictors.columns:
        if not df[col].isnull().any():
            cols_with_no_nulls.append(col)
    return cols_with_no_nulls
```

```python
# Get the columns that dont have any missing values
```

```python
num_columns = get_col_with_no_na(combined_data, 'num')
```

```python
cat_columns = get_col_with_no_na(combined_data, 'no_num')
```

```python
# Encode the categorial variables using OneHotEncode
def oneHotEncode(df, colnames):
    for col in colnames:
        if(df[col].dtype == np.dtype('object')):
            dummies = pd.get_dummies(df[col], prefix=col)
            df = pd.concat([df, dummies], axis = 1)
            
            # drop encoded col
            df.drop([col], axis=1, inplace = True)
    return df
```

```python
combined.shape[1]
```

```python
combined = oneHotEncode(combined, cat_columns)
```

```python
combined.shape[1]
```

```python
# Now lets split the combined df back into training and test data sets
# For this we need to know how many rows of training data were present in the original training dataset
```

```python
train_rows = train.shape[0]
```

```python
train_rows
```

```python
train_data = combined[:1460]
```

```python
test_data = combined[1460:]
```

```python
# Columns in the train_data will define the input dimensions in ann
train_data.shape[1]
```

```python
# Artificial Neural Networks 
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
pred_model = Sequential()
```

```python
# Adding Input layers and First hidden layer
# We will choose the rectifier function for input layer and use the sigmoid activation function for output
# Unit = output_dim is choosen as 128 
# input_dim is the no of input variables which can be written as train_data.shape[1]
```

```python
pred_model.add(Dense(units= 128, activation = 'relu', input_dim = train_data.shape[1], kernel_initializer = 'normal'))
```

```python
# Adding 5 Hidden layers
pred_model.add(Dense(units= 256, activation = 'relu', kernel_initializer = 'normal'))
pred_model.add(Dense(units= 256, activation = 'relu', kernel_initializer = 'normal'))
pred_model.add(Dense(units= 256, activation = 'relu', kernel_initializer = 'normal'))
pred_model.add(Dense(units= 256, activation = 'relu', kernel_initializer = 'normal'))
pred_model.add(Dense(units= 256, activation = 'relu', kernel_initializer = 'normal'))
```

```python
# Adding the output layer
pred_model.add(Dense(units= 1, activation = 'linear', kernel_initializer = 'normal'))
```

```python
# Compiling the ANN
pred_model.compile(optimizer = 'adam', loss = 'mean_absolute_error', metrics = ['mean_absolute_error'])
```

```python
pred_model.summary()
```

```python
pred_model.fit(train_data, dep_var, batch_size = 100, nb_epoch = 500, validation_split=0.2)
```

```python
# Validation loss is 21549.69
```

```python
SalePrice_pred = pred_model.predict(test_data)
```
